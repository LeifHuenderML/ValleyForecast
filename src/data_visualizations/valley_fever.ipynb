```python
import plotly.graph_objects as go
import pandas as pd
import plotly.express as px

print("Importing Libraries Complete")
```

```python
df = pd.read_csv('../../data/valley_fever_rates/valley_fever_rates_05_30_24.csv')
print(df['County'].unique())
df.head()
```

```python
## Graph 1
# creating a x y line graph with each county inserted over the span of the years tracking the year and the number of cases

counties = df['County'].unique()

traces = []
for county in counties:
    tmp = df[df['County'] == county]
    trace = go.Scatter(x=df['Year'], y=tmp['Cases'], name=county)
    traces.append(trace)


fig = go.Figure(data=traces)
fig.show()
```

```python
fig = px.line(df, x='Year', y='Cases')
fig.show()
```
