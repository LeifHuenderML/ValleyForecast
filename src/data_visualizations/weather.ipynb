```python
import plotly.graph_objects as go
import plotly.express as px
import pandas as pd
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv('../../data/weather_datasets/label_encoded_weather_06_02_24.csv')
```

```python
df.head()
```

```python
# the size is too much so i am aggregating it so that each year is a average of all the instances for that location for that year
aggregated_df = df.groupby(['County', 'Year']).mean().reset_index()

year = aggregated_df['Year']
aggregated_df = aggregated_df.drop(columns=['Year', 'Month', 'Day', 'Hour'], axis=1)

fig = px.parallel_coordinates(aggregated_df, 
                              color=year, 
                              color_continuous_scale=px.colors.diverging.Tealrose,
                              title="Weather Parameters Aggregated by Year")
fig.update_layout(
    paper_bgcolor='#e5e9f0',
    title_x = 0.5,
)
fig.show()
```
